# Gr치ficas AOD

```python
# Se importan las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from glob import glob
```

## Grafica promedio mensual AOD

```python
pathInput = './output/averages/month/'
# Se define los a침os a analizar
years = ['2018', '2019', '2020', '2021', '2022', '2023']
```

```python
# Se leen los archivos de cada a침o
files_month = []
for year in years:
    files = glob(pathInput + year + '/geotiff/*.tif')
    files_month.append(files)

print(files_month)
print(len(files_month))
```

```python
# Por cada a침o se leen los archivos de cada mes y se saca el promedio de cada mes y se guardan en una lista
monthly_means = []
for year in files_month:
    monthly_means_year = []
    for month in year:
        with rasterio.open(month) as src:
            data = src.read(1)
            data = np.where(data == -9999, np.nan, data)
            monthly_means_year.append(np.nanmean(data))
    monthly_means.append(monthly_means_year)

print(monthly_means)
print(len(monthly_means))
```

```python
# Se grafica el promedio de cada mes por a침o, en la grafica las lineas son punteadas y el eje x viene los noombres de los meses, se aumenta el tama침o de la grafica, y se agrega el titulo d elos ejes y el titulo de la grafica
fig, ax = plt.subplots()
for i in range(len(monthly_means)):
    ax.plot(monthly_means[i], label=years[i], linestyle='--')
ax.set_xticks(np.arange(0, 12))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
fig.set_size_inches(10, 6)
plt.xlabel('Month')
plt.ylabel('Average AOD GOES-16/ABI')
plt.title('Average AOD GOES-16/ABI ZMVM 2018-2023')
plt.legend()
plt.show()

# Se guarda la grafica en un archivo png
fig.savefig('./output/averages/month/average_monthly_aod.png')
```

## Gr치fica promedio anual

```python
pathInput = './output/averages/year/'
# Se define los a침os a analizar
years = ['2018', '2019', '2020', '2021', '2022', '2023']
```

```python
# Se leen los archivos de cada a침o
files_year = []
for year in years:
    files = glob(pathInput + year + '/geotiff/*.tif')
    files_year.append(files)

print(files_year)
print(len(files_year))
```

```python
# Se leen los archivos de cada a침o y se saca el promedio de cada a침o y se guardan en una lista
yearly_means = []
for year in files_year:
    yearly_means_year = []
    for month in year:
        with rasterio.open(month) as src:
            data = src.read(1)
            data = np.where(data == -9999, np.nan, data)
            yearly_means_year.append(np.nanmean(data))
    yearly_means.append(yearly_means_year)

print(yearly_means)
print(len(yearly_means))
```

```python
# Convierte la lista de listas en una lista
yearly_means = [item for sublist in yearly_means for item in sublist]

# Se grafica el promedio de cada a침o, en una grafcia de barras, se aumenta el tama침o de la grafica, y se agrega el titulo d elos ejes y el titulo de la grafica y se rotan los nombres de los a침os en el eje x
fig, ax = plt.subplots()
ax.bar(years, yearly_means)
fig.set_size_inches(10, 6)
plt.xlabel('Year')
plt.ylabel('Average AOD GOES-16/ABI')
plt.title('Average AOD GOES-16/ABI ZMVM 2018-2023')
plt.xticks(rotation=45)
plt.show()

# Se guarda la grafica en un archivo png
fig.savefig('./output/averages/year/average_yearly_aod.png')

```
